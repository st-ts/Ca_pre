```python
def count_licks_in_trial(row, data_licks):
    # Get licks within the start and end time of the trial
    licks_in_trial = len(data_licks[(data_licks[0] >= row['resp_start_ts']) & 
                                 (data_licks[0] <= row['resp_start_ts'] + 1500)]) 
    return licks_in_trial
```

```python
file_path = fc.selected_path
data_all = {}
dir = os.listdir(file_path)
files = []
for file in dir:
    if file[-6:]=='B2.csv':
        files.append(file)
        file_path = fc.selected_path + '\\' + file
        data = pd.read_csv(file_path)
        # trim the data
        data = data[data['laser_ts']!=0]
        # licks timing info
        licks_L = pd.read_csv(file_path[0:-4] + '_t_licks_L.csv', header = None)
        licks_R = pd.read_csv(file_path[0:-4] + '_t_licks_R.csv', header = None)
        # add a column counting the licks in the trials
        data['num_licks_L'] = data.apply(lambda row: count_licks_in_trial(row, licks_L), axis=1)
        data['num_licks_R'] = data.apply(lambda row: count_licks_in_trial(row, licks_R), axis=1)
        # add data for previous choice
        prev_choice = [0]
        prev_choice.extend(data['choice'][:-1].tolist())
        data['prev_choice'] = prev_choice
        # add 1st choice column
        n_tr = data.shape[0]
        first_choice = [0]* n_tr
        first_choice_ts = [float("nan")]*n_tr
        frame_first_choice = [float("nan")]*n_tr
        for n in np.arange(n_tr):
            t=data['resp_start_ts'][n]
            ll = [l for l in licks_L if ( (l>t) & (l<t+1500) )]
            lr = [l for l in licks_R if ( (l>t) & (l<t+1500) )]
            if (len(ll) >  0) | (len(lr) > 0):
                if len(lr)==0:
                    first_choice[n]=1
                    first_choice_ts[n]=ll[0]
                    frame_first_choice[n]=round(ll[0]/100)
                elif len(ll)==0:
                    first_choice[n]=-1
                    first_choice_ts[n]=lr[0]
                    frame_first_choice[n]=round(lr[0]/100)
                elif ll[0]<lr[0]:
                    first_choice[n]=1
                    first_choice_ts[n]=ll[0]
                    frame_first_choice[n]=round(ll[0]/100)
                else:
                    first_choice[n]=-1
                    first_choice_ts[n]=lr[0]
                    frame_first_choice[n]=round(lr[0]/100)

        data['first_choice'] = first_choice
        data[data['first_choice'] != 0]
        ## final chord
        # add it all the the data_all
        exp = {'data':data, 'licks_L':licks_L, 'licks_R':licks_R}
        data_all[file[:4]] = exp
```

```python
# visualise the missed trials fro different laser types for all mice
n_exp = len(files)
cols = 3
rows = math.ceil(n_exp/cols)
fig, axs = plt.subplots(rows, cols)
axn=0
laser_tags = ['no', 'left', 'right']
exps = list(data_all.keys())
n=0
for ax in axs.flatten():
    data=data_all[exps[n]]['data']
    
    data_las_left = data[(data['laser'] == 1)]
    data_las_right = data[data['laser'] == -1]
    data_las_no = data[data['laser'] == 0]
    missed = [np.mean(data_las_no['missed']),
              np.mean(data_las_left['missed']),
              np.mean(data_las_right['missed'])]
    ax.bar(laser_tags,missed)
    ax.set_title(exps[n])
    n=n+1

fig.suptitle('Missed trials for laser conditions') 
plt.tight_layout()
plt.show()

```
